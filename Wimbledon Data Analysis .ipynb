# The Championships, Wimbledon Records (Final) Open Era 1968-2020

The Championships, Wimbledon, commonly known simply as Wimbledon or The Championships, is the oldest tennis tournament in the world and is regarded by many as the most prestigious.It has been held at the All England Club in Wimbledon, London, since 1877 and is played on outdoor grass courts, and since 2009 with a retractable roof over Centre Court, and since 2019, over No. 1 Court.
Due to the COVID-19 pandemic, the 2020 edition of the tournament was cancelled for the first time since World War II.

Until 1922, the reigning champion had to play only in the final, against whomever had won through to challenge him/her. As with the other three Major or Grand Slam events, Wimbledon was contested by top-ranked amateur players; professional players were prohibited from participating. This changed with the advent of the open era in 1968.

# Most Championships by a Single Player

```python
Most_Champions=Wimbledon.Champion.value_counts()
Most_Champions
```

Roger Federer has won record 8 Wimbledon Championships

```python
Most_Champions.plot(kind='bar',rot=90,figsize=(11,5),title='Most No of Championships in Open era')
```

```python
Most_Finals1=Wimbledon.Champion.value_counts()
Most_Finals1.head()
```

```python
Most_Finals2=Wimbledon.Runnerup.value_counts()
Most_Finals2.head()
```

```python
Most_Finals=pd.merge(Most_Finals1,Most_Finals2,right_index=True,left_index=True)
Most_Finals
```

Most Number of Appearance in Finals of the Championships

```python
Most_Finals['Total']=Most_Finals.Champion + Most_Finals.Runnerup
Most_Finals
```

# Roger Federer has cumulative highest no of appearance in finals

```python
Most_Finals.Total.plot(kind='bar',figsize=(11,5),title='Most no of Appearance in Finals')
```
